```python
# imports
import pandas as pd
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import csv
# read in excel file
df = pd.read_excel("PubChem IDs.xlsx")



csv_file = "result.csv"


# print(df.head())

compound_list = df["PubChem ID"]
smiles_list = []
list_of_compounds_done = []
list_of_compounds_not_done = []
list_of_compounds_timeout =[]

with open(csv_file, mode="w", newline="") as file:
    pass 

# funcrion to add s,iles ot csv file
def check_existing_data(data):
    with open(csv_file, mode="r") as file:
        reader = csv.DictReader(file)
        for row in reader:
            if row["ID"] == data["ID"] and row["name"] == data["name"]:
                return True
    return False

def add_item_to_csv(item):
        with open(csv_file, mode="a", newline="") as file:
            fieldnames = ["ID", "name"]
            writer = csv.DictWriter(file, fieldnames=fieldnames)

            # Check if the file is empty, and if so, write the header
            if file.tell() == 0:
                writer.writeheader()

            # Write the item to the CSV file
            writer.writerow(item)


for compound in tqdm(compound_list, desc="Fetching SMILES"):
    if compound not in list_of_compounds_done:
        if compound is str:
            url_template = (
                f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/name/{compound}/JSON"
            )
        else:
            url_template = (
                f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/cid/{compound}/JSON"
            )
        response = requests.get(url_template)
        try:
            smiles = response.json()["PC_Compounds"][0]["props"][19]["value"]["sval"]
            item = {"ID": compound, "name": smiles}
            add_item_to_csv(item)

        except (KeyError, IndexError):
            try:
                smiles = response.json()["PC_Compounds"][0]["props"][18]["value"]["sval"]
            except KeyError:
                smiles = "N/A"  # If both keys are not found
            except IndexError:
                smiles="N/A"

    smiles_list.append(smiles)


    
```
