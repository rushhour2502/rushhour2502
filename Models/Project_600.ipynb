```python

# Ensure PUBCHEM_COMPOUND_CID is consistent across all DataFrames
df1 = df[['PUBCHEM_COMPOUND_CID', 'docking score']].copy()
df2 = df_2[['PUBCHEM_COMPOUND_CID', 'docking score']].copy()
df3 = df_3[['PUBCHEM_COMPOUND_CID', 'docking score']].copy()
df4 = df_4[['PUBCHEM_COMPOUND_CID', 'docking score']].copy()

# Convert PUBCHEM_COMPOUND_CID to str (or int if applicable)
df1['PUBCHEM_COMPOUND_CID'] = df1['PUBCHEM_COMPOUND_CID'].astype(str)
df2['PUBCHEM_COMPOUND_CID'] = df2['PUBCHEM_COMPOUND_CID'].astype(str)
df3['PUBCHEM_COMPOUND_CID'] = df3['PUBCHEM_COMPOUND_CID'].astype(str)
df4['PUBCHEM_COMPOUND_CID'] = df4['PUBCHEM_COMPOUND_CID'].astype(str)

# Rename docking score columns before merging
df1 = df1.rename(columns={'docking score': 'docking_score_7cx2'})
df2 = df2.rename(columns={'docking score': 'docking_score_7ymj'})
df3 = df3.rename(columns={'docking score': 'docking_score_7yrc'})
df4 = df4.rename(columns={'docking score': 'docking_score_1pk2'})

# Merge all DataFrames on 'PUBCHEM_COMPOUND_CID'
df_new = df1.merge(df2, on='PUBCHEM_COMPOUND_CID', how='outer') \
            .merge(df3, on='PUBCHEM_COMPOUND_CID', how='outer') \
            .merge(df4, on='PUBCHEM_COMPOUND_CID', how='outer')

# Display the first few rows

```
