```python
import numpy as np 

def pct_change(df, columns, new_column):
    previous_value_holder = np.nan
    new_column_holder = []
    for value in df[columns]:
        if value is None:
            new_column_holder.append(np.nan)
        else:
            new_column_holder.append((value - previous_value_holder)/previous_value_holder)
        previous_value_holder = value 
        
    df[new_column] = new_column_holder 
    return df.head()
```

```python
data = pd.DataFrame({
    "Price": [100, 102, 105, 103, 108]
})

result = pct_change(data, "Price", "Change")
print(result)
```

```python
class Employee:
    number_of_employees = 0
    raise_amount = 1.04
    def __init__(self, first, last, salary):
        self.first = first
        self.last = last
        self.salary = salary
        self.email = first + '.' + last + '.@company.com'
        Employee.number_of_employees += 1
    
    def fullname(self):
        return "{} {}".format(self.first, self.last)
    
    def apply_raise(self):
        self.salary = self.salary * Employee.raise_amount
    
emp1 = Employee("Steve", "Johanson", 30000)
emp2 = Employee("Jerry", "Maguire", 50000)

print (Employee.number_of_employees)

```
