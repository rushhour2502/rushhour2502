```python
import calendar
import matplotlib.pyplot as plt

# Create a figure with 3 columns and 4 rows
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 10))

# Iterate over each month and place it in the corresponding subplot
for month in range(1, 13):
    ax = axes[(month-1)//3, (month-1)%3]  # Get row and column index
    ax.text(0.5, 0.5, calendar.month(2025, month), fontsize=12, ha='left', va='center', family='monospace')
    ax.axis('off')  # Hide axes

# Adjust layout and display the grid of calendars
plt.tight_layout()
plt.show()
```

MY HANGMAN GAME

```python
import random 

def hangman():
    word = random.choice(['man', "boy", "chair", "dog"])
    validinputs = "abcdefghijklmnopqrstuvwxyz"
    turns = 10
    guessmade = ''

    while len(word) > 0:
        main = ''
        missed = 0

        for letter in word:
            if letter in guessmade:
                main+= letter
                missed += 1
            else:
                main+= "_" + ' '
        if main == word:
            print(word)
            print("You win!")
            break
        print ("Guess the word ", main)
        guess = input()

        if guess in validinputs:
            guessmade += guess
        else:
            print("Enter a valid character")
            guess = input()

        if guess not in word:
            print(f"You have {turns} tries left")
            print("----------------------------")
            turns -= 1
            if turns == 1:
                print ("last try, think carefully")
            if turns == 0:
                print("You loose")
                print("You let a kind man die")
                break
        


name = input("Enter your name: ")
print("Welcome ", name)
print("Try to guess the right word in less than 10 attempts")
print("---------------")
hangman()

```

```python
def magicBall():
  import random
  answers = ['Yes - definitely.', 'It is decidedly so.', 'Without a doubt.', 'Reply hazy, try again.', 'Ask again later.', 'Better not tell you now.', 'My sources say no.', 'Outlook not so good.', 'Very doubtful.']
  ln = len(answers)
  while True:
    Question = input("What do yo want to know about: ")
    if len(Question) > 4:
        select = random.randrange(0, ln)
        reply = answers[select]
        return (
        f'''
    Question: {Question}
    Answer: {reply}
    '''
        )
    else:
        print("Ask a valid question dear")
        
print(magicBall())

```
